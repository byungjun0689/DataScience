# FizzBuzz 
 - 3의 배수 Fizz
 - 5의 배수 Buzz
 - 3/5 의 공배수 Fizz Buzz

```python
for i in range(1,31):
    if (i % 3)== 0 and (i % 5) == 0 :
        print("FizzBuzz")
    elif (i % 3) == 0 and (i % 5) != 0:
        print("Fizz")
    elif (i % 3) != 0 and (i % 5) == 0:
        print("Buzz")
    else:
        print(i)
```
