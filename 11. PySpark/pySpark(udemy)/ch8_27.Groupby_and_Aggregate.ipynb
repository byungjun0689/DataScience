# Spark DataFrames
## Groupby and Aggregate Functions

```python
import findspark
```

```python
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7/')
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('aggs').getOrCreate()
```

```python
df = spark.read.csv('/home/ubuntu/DataScience/pySpark/Spark_DataFrames/sales_info.csv', inferSchema=True, header=True)
```

```python
df.show()
```

```python
df.printSchema()
```

### Groupby

```python
df.groupBy('Company')
```

```python
df.groupBy('Company').mean() # get DataFrame
```

```python
df.groupBy('Company').mean().show()
```

```python
df.groupBy('Company').max().show()
```

```python
df.groupBy('Company').min().show()
```

```python
df.groupBy('Company').count().show()
```

```python
df.agg({'Sales':'sum'}).show() # column, function
```

```python
df.agg({'Sales':'max'}).show() # column, function
```

```python
group_data = df.groupBy('Company')
```

```python
group_data.agg({'Sales':'max'}).show()
```

```python
group_data.agg({'Sales':'mean'}).show()
```

```python
from pyspark.sql.functions import countDistinct, avg, stddev
```

```python
df.select(countDistinct('Sales')).show()
```

```python
df.select(avg('Sales')).show()
```

```python
df.select(stddev('Sales')).show()
```

```python
from pyspark.sql.functions import format_number
```

```python
sales_std = df.select(stddev('Sales').alias('std'))
```

```python
sales_std.show()
```

```python
sales_std.select(format_number('std',2)).show()
```

```python
df.orderBy('Sales').show()
```

```python
df.orderBy(df['Sales'].desc()).show()
```
