#### Gray tone  

```python
import random

def gray_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60,100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from matplotlib import font_manager, rc
```

```python
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
from konlpy.corpus import kolaw
from konlpy.tag import Komoran
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tagger = Komoran() 
```

```python
def get_noun(text):
    nouns = tagger.nouns(text)
    return [n for n in nouns if len(n) > 1] # 2 
```

```python
ko_con_text = kolaw.open("constitution.txt").read()
ko_con_text[:100]
```

```python
cv = CountVectorizer(tokenizer=get_noun)
```

```python
tdm = cv.fit_transform(ko_con_text.split("\n"))
```

```python
tdm
```

```python
words = cv.get_feature_names()
```

```python
count_mat = tdm.sum(axis=0)
count = np.squeeze(np.asarray(count_mat))
```

```python
word_count = list(zip(words,count))
```

```python
word_count = sorted(word_count, key=lambda x:x[1], reverse=True)
```

```python
word_count[:20]
```

```python
korea_coloring = np.array(Image.open('data/korea_mask.jpg'))
```

```python
image_colors = ImageColorGenerator(korea_coloring)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(korea_coloring, cmap = plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(font_path='C:\\Windows\\Fonts\\malgun.ttf', background_color='white', width=300, height=200, max_words=2000, 
               stopwords=stopwords, max_font_size=40, random_state=42)
cloud = wc.generate_from_frequencies(dict(word_count))
plt.figure(figsize=(10,7))
plt.imshow(cloud)
plt.axis("off")
```

```python
wc = WordCloud(font_path='C:\\Windows\\Fonts\\malgun.ttf', background_color='white',mask=korea_coloring, 
               min_font_size=1, max_font_size=40, random_state=42, relative_scaling=0.2)
cloud = wc.generate_from_frequencies(dict(word_count))
plt.figure(figsize=(12,12))
plt.imshow(cloud.recolor(color_func=image_colors),interpolation='bilinear')
plt.axis("off")
plt.show()
```
